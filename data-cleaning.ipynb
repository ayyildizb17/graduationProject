```python
import pandas as pd
import csv


studentInfoData = pd.read_csv("../anonymisedData/studentInfo.csv")
regions = studentInfoData['region'].unique()
irelandData1 = []
southRegionData1 = []
northRegionData1 = []
westMidlandsData1 = []
southWestRegion1 = []
northWesternRegion1 = []
yorkshireRegion1 = []
scotland1 = []

firstPartArrayIreland = []
secondPartArrayIreland = []
firstPartArrayNorthRegion = []
secondPartArrayNorthRegion = []
firstPartArrayWestMidlands = []
secondPartArrayWestMidlands = []
firstPartArraySouthRegion = []
secondPartArraySouthRegion = []
firstPartArraySouthWestRegion = []
secondPartArraySouthWestRegion = []
firstPartArrayNorthWesternRegion = []
secondPartArrayNorthWesternRegion = []
firstPartArrayYorkshireRegion = []
secondPartArrayYorkshireRegion = []
firstPartArrayScotland = []
secondPartArrayScotland = []

irelandData =  studentInfoData[studentInfoData['region'] == 'Ireland']
nonNaIrelandData = irelandData.dropna(subset=['imd_band'])
irelandData1.append(nonNaIrelandData["imd_band"].values)

northRegionData = studentInfoData[studentInfoData['region'] == 'North Region']  
nonNaNorthRegionData = northRegionData.dropna(subset=['imd_band'])
northRegionData1.append(nonNaNorthRegionData["imd_band"].values)

westMidlandsData = studentInfoData[studentInfoData['region'] == 'West Midlands Region']
nonNaWestMidlandsData = westMidlandsData.dropna(subset=['imd_band'])
westMidlandsData1.append(nonNaWestMidlandsData["imd_band"].values)

southRegionData = studentInfoData[studentInfoData['region'] == 'South Region']
nonNaSouthRegionData = southRegionData.dropna(subset=['imd_band'])
southRegionData1.append(nonNaSouthRegionData["imd_band"].values)

southWestRegionData = studentInfoData[studentInfoData['region'] == 'South West Region']
nonNaSouthWestRegionData = southWestRegionData.dropna(subset=['imd_band'])
southWestRegion1.append(nonNaSouthWestRegionData["imd_band"].values)

northWesternRegionData = studentInfoData[studentInfoData['region'] == 'North Western Region']
nonNaNorthWesternRegionData = northWesternRegionData.dropna(subset=['imd_band'])
northWesternRegion1.append(nonNaNorthWesternRegionData["imd_band"].values)

yorkShireRegionData = studentInfoData[studentInfoData['region'] == 'Yorkshire Region']
nonNaYorkShireRegionData = yorkShireRegionData.dropna(subset=['imd_band'])
yorkshireRegion1.append(nonNaYorkShireRegionData["imd_band"].values)

scotlandData = studentInfoData[studentInfoData['region'] == 'Scotland']
nonNaScotlandData = scotlandData.dropna(subset=['imd_band'])
scotland1.append(nonNaScotlandData["imd_band"].values)
irelandData["imd_band"] = irelandData["imd_band"].astype(str)
for i in range(0,len(irelandData1[0])):
    firstPartArrayIreland.append(int((irelandData1[0][i]).split("-")[0]))
    secondPartArrayIreland.append(int((irelandData1[0][i]).split("-")[1].split("%")[0]))


  
for i in range(0,len(northRegionData1[0])):
    firstPartArrayNorthRegion.append(int((northRegionData1[0][i]).split("-")[0]))
    secondPartArrayNorthRegion.append(int((northRegionData1[0][i]).split("-")[1].split("%")[0]))
    
for i in range(0,len(westMidlandsData1[0])):
    firstPartArrayWestMidlands.append(int((westMidlandsData1[0][i]).split("-")[0]))
    secondPartArrayWestMidlands.append(int((westMidlandsData1[0][i]).split("-")[1].split("%")[0]))    

for i in range(0,len(southRegionData1[0])):
    firstPartArraySouthRegion.append(int((southRegionData1[0][i]).split("-")[0]))
    secondPartArraySouthRegion.append(int((southRegionData1[0][i]).split("-")[1].split("%")[0]))
        
for i in range(0,len(southWestRegion1[0])):
    firstPartArraySouthWestRegion.append(int((southWestRegion1[0][i]).split("-")[0]))
    secondPartArraySouthWestRegion.append(int((southWestRegion1[0][i]).split("-")[1].split("%")[0]))
    
for i in range(0,len(northWesternRegion1[0])):
    firstPartArrayNorthWesternRegion.append(int((northWesternRegion1[0][i]).split("-")[0]))
    secondPartArrayNorthWesternRegion.append(int((northWesternRegion1[0][i]).split("-")[1].split("%")[0]))
    
for i in range(0,len(yorkshireRegion1[0])):
    firstPartArrayYorkshireRegion.append(int((yorkshireRegion1[0][i]).split("-")[0]))
    secondPartArrayYorkshireRegion.append(int((yorkshireRegion1[0][i]).split("-")[1].split("%")[0]))    

for i in range(0,len(scotland1[0])):
    firstPartArrayScotland.append(int((scotland1[0][i]).split("-")[0]))
    secondPartArrayScotland.append(int((scotland1[0][i]).split("-")[1].split("%")[0]))    
    
meanIreland = str(int(sum(firstPartArrayIreland) / len(firstPartArrayIreland))) + "-" + str(int(sum(secondPartArrayIreland) / len(secondPartArrayIreland))) + "%"

for i in range(len(irelandData["imd_band"])):
    if str(irelandData["imd_band"].iloc[i]) == "nan":
        irelandData["imd_band"].iloc[i] = str(meanIreland)
    if "%" not in str(irelandData["imd_band"].iloc[i]):
    
        print(irelandData["imd_band"].iloc[i])
        irelandData["imd_band"].iloc[i] = irelandData['imd_band'].iloc[i] + "%"
    
studentInfoData.update(irelandData)


meanNorthRegion = str(int(sum(firstPartArrayNorthRegion) / len(firstPartArrayNorthRegion))) + "-" + str(int(sum(secondPartArrayNorthRegion) / len(secondPartArrayNorthRegion))) + "%"
for i in range(len(northRegionData["imd_band"])):
    if str(northRegionData["imd_band"].iloc[i]) == "nan":
        northRegionData["imd_band"].iloc[i] = str(meanNorthRegion)
    if "%" not in str(northRegionData["imd_band"].iloc[i]):
    
        print(northRegionData["imd_band"].iloc[i])
        northRegionData["imd_band"].iloc[i] = northRegionData['imd_band'].iloc[i] + "%"
studentInfoData.update(northRegionData)

meanWestMidlans = str(int(sum(firstPartArrayWestMidlands) / len(firstPartArrayWestMidlands))) + "-" + str(int(sum(secondPartArrayWestMidlands) / len(secondPartArrayWestMidlands))) + "%"
for i in range(len(westMidlandsData["imd_band"])):
    if str(westMidlandsData["imd_band"].iloc[i]) == "nan":
        westMidlandsData["imd_band"].iloc[i] = str(meanWestMidlans)
    if "%" not in str(westMidlandsData["imd_band"].iloc[i]):
    
        print(westMidlandsData["imd_band"].iloc[i])
        westMidlandsData["imd_band"].iloc[i] = westMidlandsData['imd_band'].iloc[i] + "%"
studentInfoData.update(westMidlandsData)

meanSouthRegion = str(int(sum(firstPartArraySouthRegion) / len(firstPartArraySouthRegion))) + "-" + str(int(sum(secondPartArraySouthRegion) / len(secondPartArraySouthRegion))) + "%"
for i in range(len(southRegionData["imd_band"])):
    if str(southRegionData["imd_band"].iloc[i]) == "nan":
        southRegionData["imd_band"].iloc[i] = str(meanSouthRegion)
    if "%" not in str(southRegionData["imd_band"].iloc[i]):
    
        print(southRegionData["imd_band"].iloc[i])
        southRegionData["imd_band"].iloc[i] = southRegionData['imd_band'].iloc[i] + "%"
studentInfoData.update(southRegionData)

meanSouthWestRegion = str(int(sum(firstPartArraySouthWestRegion) / len(firstPartArraySouthWestRegion))) + "-" + str(int(sum(secondPartArraySouthWestRegion) / len(secondPartArraySouthWestRegion))) + "%"
for i in range(len(southWestRegionData["imd_band"])):
    if str(southWestRegionData["imd_band"].iloc[i]) == "nan":
        southWestRegionData["imd_band"].iloc[i] = str(meanSouthWestRegion)
    if "%" not in str(southWestRegionData["imd_band"].iloc[i]):
    
        print(southWestRegionData["imd_band"].iloc[i])
        southWestRegionData["imd_band"].iloc[i] = southWestRegionData['imd_band'].iloc[i] + "%"
studentInfoData.update(southWestRegionData)

meanNorthWesternRegion = str(int(sum(firstPartArrayNorthWesternRegion) / len(firstPartArrayNorthWesternRegion))) + "-" + str(int(sum(secondPartArrayNorthWesternRegion) / len(secondPartArrayNorthWesternRegion))) + "%"
for i in range(len(northWesternRegionData["imd_band"])):
    if str(northWesternRegionData["imd_band"].iloc[i]) == "nan":
        northWesternRegionData["imd_band"].iloc[i] = str(meanNorthWesternRegion)
    if "%" not in str(northWesternRegionData["imd_band"].iloc[i]):
    
        print(northWesternRegionData["imd_band"].iloc[i])
        northWesternRegionData["imd_band"].iloc[i] = northWesternRegionData['imd_band'].iloc[i] + "%"
studentInfoData.update(northWesternRegionData)

meanYorkshireRegion = str(int(sum(firstPartArrayYorkshireRegion) / len(firstPartArrayYorkshireRegion))) + "-" + str(int(sum(secondPartArrayYorkshireRegion) / len(secondPartArrayYorkshireRegion))) + "%"
for i in range(len(yorkShireRegionData["imd_band"])):
    if str(yorkShireRegionData["imd_band"].iloc[i]) == "nan":
        yorkShireRegionData["imd_band"].iloc[i] = str(meanYorkshireRegion)
    if "%" not in str(yorkShireRegionData["imd_band"].iloc[i]):
    
        print(yorkShireRegionData["imd_band"].iloc[i])
        yorkShireRegionData["imd_band"].iloc[i] = yorkShireRegionData['imd_band'].iloc[i] + "%"
studentInfoData.update(yorkShireRegionData)

meanScotland = str(int(sum(firstPartArrayScotland) / len(firstPartArrayScotland))) + "-" + str(int(sum(secondPartArrayScotland) / len(secondPartArrayScotland))) + "%"
for i in range(len(scotlandData["imd_band"])):
    if str(scotlandData["imd_band"].iloc[i]) == "nan":
        scotlandData["imd_band"].iloc[i] = str(meanScotland)
    if "%" not in str(scotlandData["imd_band"].iloc[i]):
    
        print(scotlandData["imd_band"].iloc[i])
        scotlandData["imd_band"].iloc[i] = scotlandData['imd_band'].iloc[i] + "%"
studentInfoData.update(scotlandData)

studentInfoData.to_csv('cleanedData.csv', index=False)      
```
